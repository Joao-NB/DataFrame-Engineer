```python
import pandas as pd

# 📄 Leitura das planilhas CSV e carregamento em DataFrames
# As planilhas 'planilha1.csv' e 'planilha2.csv' devem estar no diretório 'arquivos'
planilha1 = pd.read_csv('data/db_fp.csv')  
planilha2 = pd.read_csv('data/db_sce.csv')  

# 📝 Renomeando a coluna 'ID. Funcional' para 'ID_FUNCIONAL' na planilha1
# Isso garante consistência no nome das colunas e permite realizar o merge corretamente
planilha1.rename(columns={'ID. Funcional': 'ID_FUNCIONAL'}, inplace=True)  

# 🔗 Realizando o merge (junção) entre as duas planilhas com base na coluna 'ID_FUNCIONAL'
# - 'how="outer"' inclui todos os registros das duas planilhas
# - 'indicator=True' adiciona a coluna '_merge' para indicar a origem das linhas:
#   ➡️ 'both': ID presente em ambas as planilhas  
#   ➡️ 'left_only': ID presente apenas na Planilha 1  
#   ➡️ 'right_only': ID presente apenas na Planilha 2  
df_merged = pd.merge(planilha1, planilha2, on='ID_FUNCIONAL', how='outer', indicator=True)  

# 🔎 Filtrando os IDs que não possuem valores correspondentes em nenhuma das duas planilhas
# - IDs apenas na Planilha 1: '_merge' == 'left_only'
# - IDs apenas na Planilha 2: '_merge' == 'right_only'
# - Mantém todas as colunas e valores originais das respectivas planilhas
ids_somente_planilha1 = df_merged[df_merged['_merge'] == 'left_only']  
ids_somente_planilha2 = df_merged[df_merged['_merge'] == 'right_only']  

# 📄 Exibindo os resultados com todas as colunas e informações das planilhas originais
print("📄 IDs que estão apenas na db_fp (com suas informações):")
display(ids_somente_planilha1)  

print("\n📄 IDs que estão apenas na db_sce (com suas informações):")
display(ids_somente_planilha2)  
```

```python
import pandas as pd

# 📂 Função para carregar planilha com verificação de existência e tratamento de exceções
def carregar_planilha(nome_arquivo):
    if not os.path.exists(nome_arquivo):
        print(f"🚫 Arquivo '{nome_arquivo}' não encontrado.")
        return None

    try:
        if nome_arquivo.endswith('.csv'):
            return pd.read_csv(nome_arquivo)
        elif nome_arquivo.endswith(('.xlsx', '.xls')):
            return pd.read_excel(nome_arquivo)
        else:
            print(f"⚠️ Formato de arquivo não suportado: {nome_arquivo}")
            return None
    except Exception as e:
        print(f"❌ Erro ao carregar '{nome_arquivo}': {e}")
        return None

# 📝 Carregando as planilhas
planilha1 = pd.read_csv('data/db_fp.csv')  
planilha2 = pd.read_csv('data/db_sce.csv')  

# 🚦 Verificando se as planilhas foram carregadas com sucesso
if planilha1 is None or planilha2 is None:
    print("🔔 Verifique os nomes e a localização dos arquivos.")
else:
    planilha1.columns = planilha1.columns.str.strip()
    planilha2.columns = planilha2.columns.str.strip()

    # 🔄 Renomeando a coluna de ID na primeira planilha para coincidir com a segunda, se necessário
    if 'ID. Funcional' in planilha1.columns and 'ID_FUNCIONAL' not in planilha1.columns:
        planilha1.rename(columns={'ID. Funcional': 'ID_FUNCIONAL'}, inplace=True)

    # ✅ Verificando se a coluna 'ID_FUNCIONAL' existe em ambas as planilhas
    if 'ID_FUNCIONAL' not in planilha1.columns or 'ID_FUNCIONAL' not in planilha2.columns:
        print("🚨 A coluna 'ID_FUNCIONAL' não foi encontrada em uma ou ambas as planilhas.")
        print(f"🔎 Colunas na planilha 1: {planilha1.columns.tolist()}")
        print(f"🔎 Colunas na planilha 2: {planilha2.columns.tolist()}")
    else:
        # 🔄 Realizando o merge com sufixos para diferenciar as colunas
        df_merged = pd.merge(
            planilha1,
            planilha2,
            on='ID_FUNCIONAL',
            how='inner',
            suffixes=('_planilha1', '_planilha2')
        )

        print("🔎 Colunas após o merge:", df_merged.columns.tolist())

        # 🧪 Identificando divergências entre colunas (exceto 'ID_FUNCIONAL')
        colunas_comuns = [col for col in planilha1.columns if col != 'ID_FUNCIONAL']

        divergencias = pd.DataFrame()
        for coluna in colunas_comuns:
            col_planilha1 = f"{coluna}_planilha1"
            col_planilha2 = f"{coluna}_planilha2"

            if col_planilha1 in df_merged.columns and col_planilha2 in df_merged.columns:
                divergencia = df_merged[df_merged[col_planilha1] != df_merged[col_planilha2]][['ID_FUNCIONAL', col_planilha1, col_planilha2]]
                if not divergencia.empty:
                    divergencia['Coluna'] = coluna  # Coluna com divergência identificada
                    divergencias = pd.concat([divergencias, divergencia])

        # 📊 Exibindo as divergências encontradas
        if not divergencias.empty:
            print("🔎 IDs com informações divergentes:")
            print(divergencias.reset_index(drop=True))
        else:
            print("✅ Nenhuma divergência encontrada entre as planilhas.")
```
