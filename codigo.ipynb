```python
import pandas as pd

# üìÑ Leitura das planilhas CSV e carregamento em DataFrames
# As planilhas 'planilha1.csv' e 'planilha2.csv' devem estar no diret√≥rio 'arquivos'
planilha1 = pd.read_csv('data/db_fp.csv')  
planilha2 = pd.read_csv('data/db_sce.csv')  

# üìù Renomeando a coluna 'ID. Funcional' para 'ID_FUNCIONAL' na planilha1
# Isso garante consist√™ncia no nome das colunas e permite realizar o merge corretamente
planilha1.rename(columns={'ID. Funcional': 'ID_FUNCIONAL'}, inplace=True)  

# üîó Realizando o merge (jun√ß√£o) entre as duas planilhas com base na coluna 'ID_FUNCIONAL'
# - 'how="outer"' inclui todos os registros das duas planilhas
# - 'indicator=True' adiciona a coluna '_merge' para indicar a origem das linhas:
#   ‚û°Ô∏è 'both': ID presente em ambas as planilhas  
#   ‚û°Ô∏è 'left_only': ID presente apenas na Planilha 1  
#   ‚û°Ô∏è 'right_only': ID presente apenas na Planilha 2  
df_merged = pd.merge(planilha1, planilha2, on='ID_FUNCIONAL', how='outer', indicator=True)  

# üîé Filtrando os IDs que n√£o possuem valores correspondentes em nenhuma das duas planilhas
# - IDs apenas na Planilha 1: '_merge' == 'left_only'
# - IDs apenas na Planilha 2: '_merge' == 'right_only'
# - Mant√©m todas as colunas e valores originais das respectivas planilhas
ids_somente_planilha1 = df_merged[df_merged['_merge'] == 'left_only']  
ids_somente_planilha2 = df_merged[df_merged['_merge'] == 'right_only']  

# üìÑ Exibindo os resultados com todas as colunas e informa√ß√µes das planilhas originais
print("üìÑ IDs que est√£o apenas na db_fp (com suas informa√ß√µes):")
display(ids_somente_planilha1)  

print("\nüìÑ IDs que est√£o apenas na db_sce (com suas informa√ß√µes):")
display(ids_somente_planilha2)  
```
